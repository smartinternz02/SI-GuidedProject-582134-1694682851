### Importing modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the dataset

```python
df = pd.read_csv('/content/House Price India.csv')
df.head()
```

### Perform the Below Visualizations: Univariate Analysis, Bi-variate Analysis, Multivariate Analysis

##### Univariate Analysis

```python
plt.plot(df['number of bathrooms'])
```

```python
plt.plot(df['living area'])
```

##### Bi-variate Analysis

```python
plt.plot(df['living area'], df['Price'], 'o')
plt.title('Price v/s Living Area')
plt.xlabel('Living Area')
plt.ylabel('Price')
```

```python
sns.boxplot(data = df, x = 'number of bedrooms', y = 'living area')
plt.title('Living Area v/s Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Living Area')
plt.show()
```

##### Multivariate analysis

```python
numerical_vars = ['number of floors', 'living area', 'lot area', 'number of bathrooms', 'Number of schools nearby', 'Distance from the airport', 'Price']
correlation_matrix = df[numerical_vars].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
